```python
import pandas as pd 
import plotly.figure_factory as ff
import plotly.express as px

# combine csvs into one dataframe from the dataset folder 
df09to19 = pd.read_csv('./Dataset/2009-2019.csv')
df20p1 = pd.read_csv('./Dataset/2020p1.csv')
df20p2 = pd.read_csv('./Dataset/2020p2.csv')
df21 = pd.read_csv('./Dataset/2021.csv')
df22 = pd.read_csv('./Dataset/2022.csv')


df = pd.concat([df09to19, df20p1, df20p2, df21, df22])

df = df[['Occur Date', 'Occur Time', 'Possible Date', 'Possible Time', 'Location', 'UCR Literal','Neighborhood', 'Latitude', 'Longitude', ]]


df.isna().sum()
```

```python
df.head()
df.dropna(subset=['Occur Date'], inplace=True)
```
